```python
# Basic Authentication system:

#create dictionary

user_credentials = {}

# function to register user




def register_user():

    username = input("Enter your username:  ")

    #check to see if username already exists

    if username in user_credentials:
      print("Username already exists. Please choose a different username.")

    else:
      password = input("Enter your password:  ")
      user_credentials[username] = password

      print("Registration successful!")

register_user()
```

```python
# The user_credentials dictionary is already defined and available from cell 0L9XQ9_KEEAG.
# The register_user function is also defined in cell 0L9XQ9_KEEAG.

# function to login user
def login_user():
    username = input("Enter your username:  ")
    password = input("Enter your password:  ")

    #check to see if username and password match
    # Accessing the global user_credentials dictionary
    if username in user_credentials and user_credentials[username] == password:
        print("Welcome Back!")
        return True # Indicate successful login
    else:
        print("Invalid username or password. Please try again.")
        return False # Indicate failed login

# Main authentication loop for interaction (using the functions)
while True:
    print("\n--- Authentication System ---")
    print("1. Register")
    print("2. Login")
    print("3. Exit")
    choice = input("Enter your choice: ")

    if choice == '1':
        register_user() # Call the register_user from cell 0L9XQ9_KEEAG
    elif choice == '2':
        if login_user():
            print("Login successful! You can now proceed with authenticated actions.")
            break # Exit the loop after successful login
    elif choice == '3':
        print("Exiting authentication system.")
        break
    else:
        print("Invalid choice. Please try again.")
```
